Il **Codice Unico Progetto (CUP)** è il “codice fiscale” di un progetto, una stringa alfanumerica di 15 caratteri che identifica univocamente un progetto di investimento pubblico finanziato con risorse pubbliche o private.

Open CUP mette a disposizione di tutti - cittadini, istituzioni ed altri enti - i dati, in formato aperto, sulle decisioni di investimento pubblico finanziate con fondi pubblici nazionali, comunitari o regionali o con risorse private registrate con il Codice Unico di Progetto. 

Importanza nei processi decisionali e di monitoraggio degli stakeholder pubblici e provati

### MIUR - Dati - Alunni per edifici scolastici

NB: i nomi colonne hanno uno space iniziale - esegui strip()

```python
# Creo il dataframe SCUOLE con le info basi delle scuole
#Carico file di anagrafica base delle scuole
fileDir = os.path.dirname(os.path.realpath('__file__'))
output_folder = os.path.join(fileDir, 'data/MIUR/anagrafica')
listing = os.listdir(output_folder)   
#loop in csv files from folder
scuole = DataFrame()
list_ = []

for infile in listing: 
    #filename = os.path.join(fileDir, infile)
    filename = os.path.join(output_folder, infile)
    filehandle = open(filename)
    print (filehandle) 
    data = pd.read_csv(filehandle, sep=",", index_col=None, header=0)
    list_.append(data)
scuole = pd.concat(list_)

scuole = scuole.rename(columns=lambda x: x.strip())
```

```python
scuole.shape
#scuole_columns = list(scuole)
scuole_columns_to_keep =[
 'AREAGEOGRAFICA',
 'CAPSCUOLA',
 'CODICECOMUNESCUOLA',
 'CODICEISTITUTORIFERIMENTO',
 'CODICESCUOLA',
 'DENOMINAZIONEISTITUTORIFERIMENTO',
 'DENOMINAZIONESCUOLA',
 'DESCRIZIONECOMUNE',
 'DESCRIZIONETIPOLOGIAGRADOISTRUZIONESCUOLA',
 'INDIRIZZOSCUOLA',
 'PROVINCIA',
 'REGIONE']
#elimino le colonne non necessarie
scuole = scuole[scuole_columns_to_keep]
scuole.shape
```

```python
scuole_cn=scuole.loc[scuole["PROVINCIA"]=="CUNEO"]
print ("Tot codici scuola univoca per provincia CUNEO: ",scuole_cn.CODICESCUOLA.nunique())
scuole_cn.groupby(["DESCRIZIONETIPOLOGIAGRADOISTRUZIONESCUOLA"]).count()
```

```python
#ISTITUTI STATALI - raggruppo per CODICESCUOLA
alunni_statali = pd.read_csv("data/MIUR/alunni/ALUCORSOETASTA20170221.csv", sep=',')

#clean columns name
alunni_statali = alunni_statali.rename(columns=lambda x: x.strip())
print ("Unique edifici statali: ", alunni_statali["CODICESCUOLA"].nunique())
alunni_statali = alunni_statali[["CODICESCUOLA", "ALUNNI"]]
alunni_statali= alunni_statali.groupby(['CODICESCUOLA']).sum()
alunni_statali["Tipo"]='S'
alunni_statali.reset_index()
```

```python
#ISTITUTI PARITARI - raggruppo per CODICESCUOLA
alunni_paritari = pd.read_csv("data/MIUR/alunni/ALUCORSOETAPAR20170221.csv", sep=',')
#clean columns name
alunni_paritari = alunni_paritari.rename(columns=lambda x: x.strip())
alunni_paritari = alunni_paritari[["CODICESCUOLA", "ALUNNI"]]
print ("Unique edifici paritari: ", alunni_paritari["CODICESCUOLA"].nunique())
alunni_paritari= alunni_paritari.groupby(['CODICESCUOLA']).sum()
alunni_paritari["Tipo"]='P'
alunni_paritari.reset_index()
```

```python
#uniamo i due dataset su CODICESCUOLA e numero alunni in uno unico
alunni= pd.concat([alunni_statali, alunni_paritari])
alunni = alunni.reset_index()
```

```python
# join scuole e alunni
print ("Iniziale: ", len(scuole_cn), "  ", len(alunni))
scuole_alunni_cn = pd.merge(scuole_cn, alunni, how="left",on='CODICESCUOLA')
print ("Finale: ",len(scuole_alunni_cn))
```

```python
scuole_alunni_cn
scuole_alunni_cn.to_csv("scuoleAlunni_cn.csv", sep='|',index=False, header=True,  encoding='utf-8')
```
